```python
import pandas as pd
import numpy as np
f_f_t = pd.read_csv("unzip/FINAL_FEATURES_TRAINTEST.tsv", delimiter = "\t")
value_not_0 = []
```

```python
len(f_f_t['i1048']) - f_f_t['i1048'].value_counts()[0]
```

```python
for i in f_f_t.columns[1:]:
    count = len(f_f_t[i]) - f_f_t[i].value_counts()[0]
    value_not_0.append(count)
```

```python
count_of_i_and_u = pd.DataFrame({'data':  value_not_0,
              'columns':  f_f_t.columns[1:]})
```

```python
count_of_i_and_u
```

```python
count_of_i_and_u = count_of_i_and_u.sort_values(by = ['data'], ascending = False)
```

```python
count_of_i_and_u
```

```python
count_of_i_and_u.to_csv('count_of_i_and_u.csv', index = False)
```
